##### Diseña un algoritmo para resolver el problema por fuerza bruta

```python
# primero se crearán 2 funciones de permutación para los números y otra para los signos operaciones:

# Función recursiva para generar permutaciones de números:
def combi_numeros(digits):
    if len(digits) == 1:
        return [digits]  # Caso base: una lista con un solo dígito tiene una permutación (ella misma)
    
    # Lista para almacenar todas las permutaciones
    permutations = []
    
    # Recorremos cada dígito
    for i in range(len(digits)):
        # Extraer el dígito actual
        current_digit = digits[i]
        
        # Generar las permutaciones de los dígitos restantes
        remaining_digits = digits[:i] + digits[i+1:]
        # Utilizando la función combi_numero recursivamente
        remaining_permutations = combi_numeros(remaining_digits)
        
        # Agregar el dígito actual a las permutaciones de los dígitos restantes
        for p in remaining_permutations:
            permutations.append([current_digit] + p)
    
    return permutations


# Función para generar el producto cartesiano de los signos operadores
def prod_signos(operators, repeat):
    # Caso base: si repeat es 1, devolver los operadores en una lista de listas
    if repeat == 1:
        return [[op] for op in operators]
    
    # Generar el producto cartesiano recursivamente
    result = []
    # Llamar recursivamente para obtener combinaciones más cortas
    subproductos = prod_signos(operators, repeat - 1)
    
    # Combinar cada operador con las combinaciones más cortas
    for op in operators:
        for subprod in subproductos:
            result.append([op] + subprod)
    
    return result

# Ejemplo de uso
##### IMPORTANTE: 
#### COMO SE EXPLICO LINEAS ARRIBA EN ESTE ALGORITMO NO SE LLEGÓ A TENER LOS RESULTADOS COMPLETOS DE CASOS PARA LA COMBINACIÓN DE DIGITOS DEL 1 AL 9
#### POR LO QUE SE APLICÓ UN CASO MÁS PEQUEÑO CONSIDERANDO SOLO LA COMBINACIÓN DE DIGITOS DEL 1 AL 6 REALIZANDO 5 OPERACIONES (REPEAT)
digits = ['1', '2', '3', '4','5','6']
operators = ['+', '-', '*', '/']
repeat = 5


# Por ultimo se crea una función para realizar todas las combinaciones y resulte con el valor Resultado_Combinación 
# tal que la diferencia con el valor objetivo sea 0

def generate_expressions_manual():
    
    # Obtener permutaciones manualmente
    digit_permutations = combi_numeros(digits)
    
    # Obtener combinaciones de operadores
    operator_combinations = prod_signos(operators,repeat)
    
    # Almacenar resultados enteros
    results = {}
    
    # Para cada permutación de cifras
    for digit_perm in digit_permutations:
        # Para cada combinación de operadores
        for op_comb in operator_combinations:
            # Construir la expresión aritmética
            expression = ''
            # el rango del for se debe a la cantidad de operaciones que se realizará, en este caso 5 debido a que tenemos 6 cifras.
            for i in range(5):
                expression += digit_perm[i] + op_comb[i]
            expression += digit_perm[5]
            
            try:
                # Evaluar la expresión
                result = eval(expression)
                
                # Solo considerar resultados enteros
                if isinstance(result, int):
                    if result not in results:
                        results[result] = []
                    results[result].append(expression)
            except ZeroDivisionError:
                # Ignorar divisiones por cero
                continue
    
    return results

# Ejecutar el algoritmo
resultados = generate_expressions_manual()

# Mostrar algunos resultados
for valor, expresiones in list(resultados.items())[:5]:  # Solo los primeros 5 resultados
    print(f"Valor: {valor}, Expresiones: {expresiones}")
```

##### (*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

El algoritmo que se planteará será un metaheuristico de Recocido Simulado.
Este algoritmo es un proceso de búsqueda por entornos basado en la aceptación probabilista de soluciones.
- El termino proviene del tratamiento de metales y procesos termodinámicos. Se realiza un enfriamiento controlado con el objetivo de proporcionar determinada estructura al metal (simulated annealing).

```python

# Función que evalúa la expresión de un individuo
def evaluar_individuo(individuo):
    try:
        # Evalúa la expresión generada por el individuo
        resultado = eval("".join(map(str, individuo)))
        return resultado
    except ZeroDivisionError:
        return float('inf')  # Penalización para divisiones por cero

# Función de aptitud (fitness)
def fun_obj(individuo):
    resultado = evaluar_individuo(individuo)
    return abs(valor_objetivo - resultado)  # Queremos minimizar la diferencia

# Generar un individuo aleatorio (secuencia alternada de dígitos y operadores)
def generar_individuo():
    individuo = []
    # el rango del for se debe a la cantidad de operaciones que se realizará, en este caso 7
    for i in range(7):  # Alternar dígitos y operadores
        if i % 2 == 0:
            individuo.append(random.choice(digitos))
        else:
            individuo.append(random.choice(operadores))
    return individuo

# Generar un vecino cercano cambiando un dígito u operador
def generar_vecino(individuo):
    vecino = individuo.copy()
    idx = random.randint(0, len(individuo) - 1)
    if idx % 2 == 0:  # Si es un dígito
        vecino[idx] = random.choice(digitos)
    else:  # Si es un operador
        vecino[idx] = random.choice(operadores)
    return vecino

# Algoritmo de recocido simulado
def recocido_simulado(temperatura_inicial, tasa_enfriamiento):
    # Inicializar la solución
    solucion_actual = generar_individuo()
    mejor_solucion = solucion_actual
    temperatura = temperatura_inicial
    
    while temperatura > .0001:
        # Generar una nueva solución vecina
        vecino = generar_vecino(solucion_actual)
        
        # Evaluar las soluciones actuales y vecinas
        fitness_actual = fun_obj(solucion_actual)
        fitness_vecino = fun_obj(vecino)
        
        # Calcular la diferencia de aptitud
        delta_fitness = fitness_vecino - fitness_actual
        
        # Aceptar el vecino si es mejor o con una probabilidad si es peor
        if delta_fitness < 0 or random.random() < math.exp(-delta_fitness / temperatura):
            solucion_actual = vecino
        
        # Actualizar la mejor solución encontrada
        if fun_obj(solucion_actual) < fun_obj(mejor_solucion):
            mejor_solucion = solucion_actual
        
        # Enfriar la temperatura
        temperatura *= tasa_enfriamiento
        
        # Imprimir el progreso
        print(f"Mejor solución: {mejor_solucion}, Fun_dif: {fun_obj(mejor_solucion)}")
        
        # Criterio de parada: si se encuentra una solución óptima
        if fun_obj(mejor_solucion) == 0:
            print("¡Se encontró una solución óptima!")
            break
    
    return mejor_solucion
```

Para calcular la complejidad del algoritmo depende de varios factores, pero se dividirá en varias partes:

- La primera parte será la primera función que se llama en recocido_simulado, la cual es generar_individuo():

- La segunda parte es la función: generar_vecino() ya que se selecciona aleatoriamente una posición en la lista de individuos y se reemplaza por un nuevo valor (un dígito o un operador). El costo total es 𝑂(1).

- La tercera parte es la función: fun_obj() y evaluar_individuo() en este caso como es un cálculo de diferencia el costo es de 𝑂(1).

- Por ultimo la función principal: recocido_simulado() en este caso El bucle principal del algoritmo se ejecuta como máximo  iteraciones_max iteraciones_max veces. Entonces es de complejidad 𝑂(iteraciones_max)

En resumen la complejidad total del algoritmo es lineal con respecto a como uno defina la máxima iteración para que aceptaría para llegar a una solución próxima.
